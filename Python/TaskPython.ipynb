### 4) Tratando a coluna da Renda

Como observado anteriormente, a **coluna da Renda possui valores ausentes**. Entretanto, esta coluna está no formato texto (strings), então não é possível trocar os valores ausentes pela média ou pela mediana por exemplo.<br><br>

Nas próximas etapas da Jornada, serão explicados diferentes jeitos de transformar variáveis categóricas em numéricas para posterior aplicação de algoritmos de Machine Learning (que exigem variáveis numéricas), mas por ora, vamos fazer o seguinte:<br>

1) Utilize o valor do salário mínimo aproximadamente como R$1000,00 (ou utilize o valor verdadeiro se quiser pesquisar) para substituir as categorias.
<br>
**Exemplo pegando valores aproximadamente na média do intervalo**:<br> 
Até 5 salários mínimos = 2500;<br>
Entre 6 e 10 salários mínimos = 7500;<br>
Entre 11 e 20 salários mínimos = 15000;<br>
Acima de 20 salários mínimos = 25000;<br>
<br>
2) Troque os valores ausentes utilizando algum desses métodos (pela média, pela mediana ou pela moda)

````python
df_clin['Renda'] = df_clin['Renda'].replace(dict(zip(['Até 5 salários mínimos','Entre 6 e 10 salários mínimos','Entre 11 a 20 salários mínimos','Acima de 21 salários mínimos'],[2500,7500,15000,25000])))
df_clin.head()
````

### 6) Tratando a coluna do USO (Começa a ficar mais difícil)

Essa coluna corresponde a uma pergunta de múltipla escolha. Entretanto a maneira como ela está apresentada não é nada visual e nem é apropriado para conseguir fazer análises sobre. Atualmente esta coluna está redigida do seguinte jeito:<br>
<br>
Resposta 1: Opção1;Opção3;Opção4<br>
Resposta 2: Opção1;Opção2<br>
Resposta 3: Opção2;Opção3;Opção4<br>
<br>
Essa coluna será tratada da seguinte maneira:<br>
1) Transformar cada resposta numa lista onde cada item é uma opção --> Exemplo: Resposta 1 = [Opção1, Opção3, Opção4] <br>
2) Criar uma nova coluna para cada tipo de opção de resposta, ou seja, terá uma coluna chamada "Opção1", outra chamada "Opção2" e assim por diante <br>
3) Agora, para cada resposta, seu valor será 1 ou 0 nas colunas que foram criadas, por exemplo:<br>
**Resposta 1: Opção1;Opção3;Opção4 --> Opção1 = 1; Opção2 = 0; Opção3 = 1; Opção4 = 1**

As opções de resposta são:<br>
<br>
1) Para armazenar meus dados médicos de forma organizada<br>
2) Para entrar em contato com meu médico, tirando possíveis dúvidas<br>
3) Para saber quais vacinas já tomei e quais ainda preciso tomar<br>
4) Para poder levar, quando necessário, resultados de exames para consultas médicas<br>
5) Não utilizaria este aplicativo de forma alguma<br>
<br>
OBS: Existia a opção da pessoa digitar alguma resposta que não estas 5, mas **não vale a pena transformar essas outras opções de respostas em novas colunas, faz mais sentido juntá-las todas numa outra coluna chamada "Outros"**<br>
<br>
6) Outros<br>
<br>
Existem diferentes maneiras de realizar essa tarefa, tente pensar em alguma e depois compare com a que foi feita no gabarito!

````python
df_clin['Uso'] = list(map(lambda x: x.split(', '), df_clin['Uso']))
df_clin.head()

usos = ['Para armazenar meus dados médicos de forma organizada','Para entrar em contato com meu médico, tirando possíveis dúvidas',
        'Para saber quais vacinas já tomei e quais ainda preciso tomar','Para poder levar, quando necessário, resultados de exames para consultas médicas',
        'Não utilizaria este aplicativo de forma alguma']

for i in range(5): #Criando colunas de opçoes e de 'outros'
    df_clin[f'Opção{i+1}'] = 0
df_clin['Outros'] = 0

for i in df_clin.index:
    for k in range(len(df_clin['Uso'][i])):
        not_in_list = True
        for j in range(len(usos)):
            if df_clin['Uso'][i][k] == usos[j]:
                df_clin[f'Opção{j+1}'][i] = 1
                not_in_list = False
                break
            else: pass
        if not_in_list: df_clin['Outros'] = 1
df_clin.head()
````

### 8) Tratando a coluna do "Quanto Pagaria" (DESAFIO)

Como vocês devem ter visto, a resposta para essa pergunta varia muito, pois ela era uma pergunta do tipo "aberta" no forms. Para conseguir analisar devidamente essa coluna, o **ideal é que ela fosse totalmente numérica**, entretanto estamos longe de atingir esse objetivo. Deletar essa coluna não é uma opção, pois ela era fundamental para o projeto.

Esse é um belo caso para vocês testarem o que aprenderam sobre manipulação avançada de strings no treinamento. Existem diversas abordagens de como fazer. Não se preocupem em conseguir extrair 100% da informação que os dados trazem para valores numéricos, mas pensem em técnicas de priorização para conseguir recuperar o máximo possível sem ter que passar resposta a resposta vendo qual o número apropriado para aquela resposta. A biblioteca regular expressions (ou regex) é muito útil para esse tipo de manipulação. Uma referência sobre a biblioteca pode ser encontrada aqui: https://www.w3schools.com/python/python_regex.asp, mas sintam-se livres para procurar outras referências também.

````python
import re
````

````python
def challenge(item):
    num =  re.findall(r'\d+',str(item) )
    if num:
        if any([word in str(item) for word in ('mes','mês','Mes','Mês','mensais')]):
            return int(num[0])*12
        else: return int(num[0])
    elif any([word in item for word in 'nao não Não Nao nada Nada'.split(' ')]): return 0
    else: return np.nan
newcol = list(map(challenge,list(df_clin['Quanto_Pagaria'])))
df_clin['Quanto_Pagaria'] = newcol
newcol = [x for x in newcol if x != np.nan]
newcol.sort()
median = newcol[len(newcol)//2]
df_clin['Quanto_Pagaria'].fillna(median,inplace=True)
````
